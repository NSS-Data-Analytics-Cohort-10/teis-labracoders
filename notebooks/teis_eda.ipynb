```python
# Dropping unnecessary columns
drop = ['Code 1', 'Code 2', 'Code 3', 'Code 4', 'Code 5', 'Code 6', 'Code 7', 'Code 8', 'Code 9', 'Code 10']
bdi_df = bdi_3.drop(drop, axis=1)
bdi_df.head()
```

```python
child_demo = bdi_df[['Child ID', 'Gender', 'Date of Birth', 'Program Label', 'Adaptive-Self Care Date of Testing']]
```

```python
child_demo.head()
```

```python
# Check distribution of males vs females in programs
child_demo['Gender'].value_counts()
```

```python
# Check evaluation distribution by type
child_demo['Program Label'].value_counts()
```

```python
# Check distinct number of children by 'Child ID'
child_demo['Child ID'].nunique()
```

```python
# Evaluations by child
assessments_bychild = child_demo.groupby('Child ID')['Program Label'].count().reset_index(name='count').sort_values('count', ascending=False).head(20)
assessments_bychild
```

```python
# General info df with child info, date of testing, & examiner
general = bdi_df.iloc[:, :5]
general['Date of Testing'] = bdi_df['Adaptive-Self Care Date of Testing']
general['Examiner'] = bdi_df['Adaptive-Self Care Examiner']
general
```

```python
bdi_totals = bdi_df.iloc[:, 45:53]
bdi_totals.head()
```

```python
totals_df = pd.concat([general, bdi_totals], axis=1)
totals_df.info()
```
