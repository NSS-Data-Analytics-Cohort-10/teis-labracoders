```python
import pandas as pd   
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
```

```python
df = pd.read_excel('../data/bdi3_all_evals_clean.xlsx')
```

```python
eligibility= df[df['Program Label'] == 'BDI-3 Eligibility Evaluation']
annual = df[df['Program Label'] == 'BDI-3 Annual Evaluation']
exit = df[df['Program Label'] == 'BDI-3 Milestone or Exit Evaluation']
```

```python
# 'BDI-3 Total Developmental Quotient	' : '#ED9924'

custom_params = {"axes.spines.right": False, "axes.spines.top": False}
sns.set_theme(style="ticks", rc=custom_params)
sns.set_theme(style="ticks", rc=custom_params)
sns.set_theme(style="ticks", rc=custom_params)
plt.figure(figsize=(10, 7)) 
sns.histplot(eligibility['BDI-3 Total Developmental Quotient'], color='#4482ca', label='Eligibility', binwidth=5)
sns.histplot(annual['BDI-3 Total Developmental Quotient'], color='#2d5e98', label='Annual', binwidth=5)
sns.histplot(exit['BDI-3 Total Developmental Quotient'], color='#1A3D66', label='Exit', binwidth=5, binrange=(40, 140))
plt.axvline(78.0,0.0,linestyle='--', color='blue', label= '25% Delay')
plt.axvline(70.0,0.0, linestyle='--',color='#a7190c', label='40% Delay')
plt.yscale('log')  
plt.xlabel('BDI-3 Total Developmental Quotient')
plt.ylabel('Log Scale')
handles, labels = plt.gca().get_legend_handles_labels() 
order = [2,3,4,1, 0] 
plt.legend([handles[i] for i in order], [labels[i] for i in order]) 
plt.savefig('../bdi_hist.png', format = 'png', bbox_inches = 'tight', dpi= 200)
plt.show()
```
